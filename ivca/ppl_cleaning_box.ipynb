```python
import pandas as pd

df = pd.read_csv('data_ppl_box.csv')

df.head(5)
```

```python
df[['box_type', 'city_abbr', 'street', 'note']] = df['name'].str.extract(r'(\w+)\s+(\w+)\s+(.*?)(?:\s+\((.*?)\))?$')
```

```python
df.to_csv('data_ppl_box_clean.csv', index=False, encoding='utf-8')
```
