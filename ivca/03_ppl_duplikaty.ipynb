```python
import pandas as pd

df = pd.read_csv('test_ppl_vse.csv')

df.head(5)
```

```python
radky_s_nan = df.isna().any(axis=1).sum()
print(f"{radky_s_nan} řádků obsahuje alespoň jednu NaN hodnotu.")
```

```python
df_nan_radky = df[df.isna().any(axis=1)]
print(df_nan_radky)
```

```python
df_clean = df.dropna()
```

```python
puvodni_pocet = len(df)
novy_pocet = len(df_clean)
print(f"Odstraněno {puvodni_pocet - novy_pocet} řádků s nulovými hodnotami.")
```

```python
df_clean.info
```

```python
df_unique = df_clean.drop_duplicates(subset=["id", "name"])
```

```python
novy_pocet = len(df_unique)
print(novy_pocet)
```
