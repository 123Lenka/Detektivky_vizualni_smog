Nejprve si nahraj data stažená z mapscrapingu ve formátu json, která jsem si pro snadnější práci u sebe ve složce pojmenovala jednoduše dataset.json a nechám si zobrazit prvních 5 řádků těchto dat.

```python
import pandas as pd

selected_clmns = pd.read_json('dataset_supermarket.json')

selected_clmns.head(5) 
```

Data se mi zdají malinko nepřehledná a nezobrazují se celá, nechám si tedy zobrazit názvy sloupců.

```python
selected_clmns.columns
```

Podle těchto názvů vyberu jen sloupce, se kterými budu chtít pracovat.

```python
selected_clmns = selected_clmns[["title", "categoryName","address","location"]]
selected_clmns.head(5)
```

Krásné, to je přesně ono. Teď jen budu potřebovat rozdělit adresu na více sloupců. Protože to neumím udělat v jednom kroku a taky se chci při práci kontrolovat, že vše vypadá jak má, tak se teď tímto budu zabývat ve více krocích.

Nejprve si rozdělím adresu podle dělítka - čárka  - tím bych měla získat adresu rosekanou na 3 část - 1. Ulice a ČP (street and number), 2. PSČ (zip_code) a Obec (city), 3. stát (country).

```python
selected_clmns[['street and number', 'zip_code and city', 'country']] = selected_clmns['address'].str.split(', ', n=2, expand=True)

selected_clmns
```

Povedlo se to, ale původní sloupec address v tabulce zůstal.
Dále rozdělím sloupec zip_code and city podle první a druhé mezery, kdy mi potom zůstane jeden sloupec s prvním částí PSČ (zip1) a druhý sloupec s druhou částí PSČ (zip 2) a nakonec střetí sloupec s názvem City.

```python
selected_clmns[['zip1', "zip2", "City"]] = selected_clmns['zip_code and city'].str.split(' ', n=2, expand=True)

selected_clmns
```

Sloupce zip1 a zip2 spojím v nový sloupec s názvem zip_code.

```python
selected_clmns['zip_code'] = selected_clmns['zip1'] + ' ' + selected_clmns['zip2']
selected_clmns.drop(columns=['zip1', 'zip2'], inplace=True)
```

```python
selected_clmns
```

Dále si rozdělím street and number podle první mezery od zadu na dva sloupce s názve street a number. Poprvé se u toho pokusím původní sloupec zahodit.

```python
selected_clmns[["street", "number"]] = selected_clmns['street and number'].str.rsplit(' ', n=1, expand=True)
selected_clmns.drop(columns=['street and number'], inplace=True)
```

```python
selected_clmns
```

Povedlo se! Ale i tak tady už je zmatek ve sloupcích, které potřebujeme a nepotřebujeme, opět si vytáhnu jen ty nejdůležitější a vypíšu prvních 5 řádků.

```python
selected_clmns = selected_clmns[["title", "categoryName","City", "street", "number", "zip_code", "location"]]
selected_clmns.head(5)
```
