Nejprve jsem si nahrála data stažená z mapscrapingu ve formátu json, která jsem si pro snadnější práci u sebe ve složce pojmenovala jednoduše dataset_supermarket.json a nechám si zobrazit prvních 5 řádků těchto dat.

```python
import pandas as pd

selected_clmns = pd.read_json('dataset_supermarket.json')

selected_clmns.head(5) 
```

Data se mi zdají malinko nepřehledná a nezobrazují se celá, nechám si tedy zobrazit názvy sloupců.

```python
selected_clmns.columns
```

Podle těchto názvů vyberu jen sloupce, se kterými budu chtít pracovat.

```python
selected_clmns = selected_clmns[["title", "categoryName","address","location"]]
selected_clmns.head(5)
```

Krásné, to je přesně ono. Teď jen budu potřebovat rozdělit adresu na více sloupců. Protože to neumím udělat v jednom kroku a taky se chci při práci kontrolovat, že vše vypadá jak má, tak se teď tímto budu zabývat ve více krocích.

Nejprve si rozdělím adresu podle dělítka - čárka  - tím bych měla získat adresu rosekanou na 3 část - 1. Ulice a ČP (street and number), 2. PSČ (zip_code) a Obec (city), 3. stát (country). Původní sloupec s názvem address zahodím.

```python
selected_clmns[['street and number', 'zip_code and city', 'country']] = selected_clmns['address'].str.split(', ', n=2, expand=True)

selected_clmns.drop(columns=['address'], inplace=True)

selected_clmns.head(5)
```

Dále rozdělím sloupec zip_code and city podle první a druhé mezery, kdy mi potom zůstane jeden sloupec s prvním částí PSČ (zip1) a druhý sloupec s druhou částí PSČ (zip 2) a nakonec třetí sloupec s názvem city. Původní sloupec zase rovnou odstraním.

```python
selected_clmns[['zip1', "zip2", "city"]] = selected_clmns['zip_code and city'].str.split(' ', n=2, expand=True)

selected_clmns.drop(columns=['zip_code and city'], inplace=True)

selected_clmns.head(5)
```

Sloupce zip1 a zip2 spojím v nový sloupec s názvem zip_code. Původní sloupce zase osdtraním.

```python
selected_clmns['zip_code'] = selected_clmns['zip1'] + ' ' + selected_clmns['zip2']
selected_clmns.drop(columns=['zip1', 'zip2'], inplace=True)
```

```python
selected_clmns.head(5)
```

Dále si rozdělím street and number podle první mezery od zadu na dva sloupce s názve street a house_number. ůvodní sloupec zase osdtraním

```python
selected_clmns[["street", "house_number"]] = selected_clmns['street and number'].str.rsplit(' ', n=1, expand=True)
selected_clmns.drop(columns=['street and number'], inplace=True)
selected_clmns.head(5)
```

Nyní se pokusím rozdělit sloupec location podle klíčů na 2 nové sloupce s názvem latitude a longitude

```python
selected_clmns['latitude'] = selected_clmns['location'].str.get('lat')
selected_clmns['longitude'] = selected_clmns['location'].str.get('lng')
selected_clmns.drop(columns=['location'], inplace=True)
selected_clmns.head(5)
```

Nyní si vyberu a přeskládám sloupce

```python
selected_clmns = selected_clmns[["title", "categoryName","city", "street", "house_number", "zip_code", "latitude", "longitude"]]
selected_clmns.head(5)
```
