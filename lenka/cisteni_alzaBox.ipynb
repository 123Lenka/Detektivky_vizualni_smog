Otevřu a prohlédnu si data set

```python
import pandas as pd


df = pd.read_excel('data_AlzaBox.xlsx')

df.columns
```

Rozdělím sloupec ulice a číslo popisné podle první mezery od zadu

```python
df[['street', 'house_number']] = df['Ulice a číslo'].str.rsplit(' ', n=1, expand=True)

df.head(5)
```

Přejmenuji si sloupce

```python
df_renamed = df.rename(columns={'Branch office': 'name',
                                'Název': 'title',
                                'Město': 'city',
                                'PSČ': 'zip_code',
                                'GeoY': 'latitude',
                                'GeoX': 'longitude',
                                'První spuštění': 'first_opening',
                                'Otevřen od': 'open_hours',
                                 'Otevřen do': 'close_hours'})

df_renamed.head(5)
```

Vyberu si jen sloupce, které potřebuji

```python
selected_clmns=df_renamed[["name","title", "city", "street", "house_number", "latitude", "longitude", "open_hours", "close_hours", "first_opening"]]
selected_clmns.head(5)
```

Zkusím uložit

```python
selected_clmns.to_excel('alzaBox_clean_verze1.xlsx',index=False)
```
