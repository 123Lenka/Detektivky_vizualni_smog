Otevření souboru s přiřazenými ruiany

```python
import pandas as pd

df = pd.read_csv('dpd_getAll_country203_kod_obce.csv')

df.head(5)
```

```python
df.columns
```

kontrola sloupců a výběr jen pár potřebných

```python
vybrane_sloupce = ['company', 'kod_obce', 'city', 'postcode', 'street', 'house_number', 'latitude', 'longitude' ]
df_vybrany = df[vybrane_sloupce]
```

```python
df_vybrany.head(5)
```

Přejmenování sloupců

```python
nove_nazvy_sloupcu = {'company': 'name',
                      'kod_obce': 'city_code',
                      'city': 'city',
                      'postcode':'zip_code',
                      'street': 'street',
                      'house_number': 'house_NO',
                      'latitude': 'latitude',
                      'longitude' :'longitude' 
                      }
df_prejmenovany = df_vybrany.rename(columns = nove_nazvy_sloupcu)
```

```python
df_prejmenovany.head(5)
```

chci zjistit kolik má můj dataset řádků

```python
df_prejmenovany.shape
```

Odebrání alzaboxu a alzawall z dat - to už neplatí, nakonec musím odstranit vše a nechat jen řádky s názvem DPD box

```python

# # Vytvoříme boolean mask, který je True pro řádky, kde 'company' obsahuje 'AlzaBox' nebo 'AlzaWall'
# mask = df_prejmenovany['name'].str.contains('AlzaBox|AlzaWall', na=False, case=False)

# # Invertujeme masku (~) a ponecháme jen řádky, které NEobsahují tyto výrazy
# df_cleaned = df_prejmenovany[~mask]


# df_cleaned.head(5)
```

```python
# Vytvoříme boolean mask, který je True pro řádky, kde 'name' obsahuje 'DPD box'
# 'na=False' zajistí, že buňky s chybějícími hodnotami (NaN) nebudou zahrnuty
# 'case=False' zajistí, že vyhledávání není citlivé na velikost písmen (najde "DPD box", "dpd box" atd.)
mask = df_prejmenovany['name'].str.contains('DPD box', na=False, case=False)

# Nyní použijeme tuto masku přímo k výběru řádků, které splňují podmínku (tj. obsahují "DPD box")
df_cleaned = df_prejmenovany[mask].copy() # Důležité: .copy() zabrání SettingWithCopyWarning


df_cleaned.head(5)
```

Opět kontrola počtem řádků z kontroly na jinych místech vím, že jich má být 243, což sedí

```python
df_cleaned.shape
```

teď si všechna name přepíšu na DPD box a nic víc za tím

```python
# Nyní v těchto vyfiltrovaných řádcích (df_cleaned) přepíšeme hodnotu sloupce 'name' na "DPD box"
# '.loc' je nejlepší způsob, jak bezpečně přistupovat a modifikovat části DataFrame
df_cleaned.loc[:, 'name'] = 'DPD box'

# Zobrazíme prvních 5 řádků, abychom zkontrolovali výsledek
df_cleaned.head(5)
```

Přidám sloupec s ID a číslováním

```python

# Použijeme index DataFramu (df_prejmenovany.index), který začíná od 0.
# Přidáme +1, abychom začali číslovat od 1.
df_cleaned['ID'] = range(1, len(df_cleaned) + 1)

# Chci, aby sloupec 'ID' byl úplně první sloupec v DataFramu,

# Nejprve získáme seznam všech sloupců
cols = df_cleaned.columns.tolist()
# Pak přesuneme 'ID' na začátek seznamu
cols = ['ID'] + [col for col in cols if col != 'ID']
# A aplikujeme nové pořadí na DataFrame
df_cleaned = df_cleaned[cols]


df_cleaned.head(5)
```

Fixně otypuju sloupce, aby nám při spojování nedělaly neplechu

```python

zmena_typu = {
    'ID': int,
    'name': str,
    'city_code': str,  
    'city': str,
    'zip_code': str,  
    'street': str,
    'house_NO': str,  
    'latitude': float,
    'longitude': float
}

df_cleaned = df_cleaned.astype(zmena_typu)
                               
df_cleaned.dtypes # Pro kontrolu datových typů
```

Uložení do nového csv

```python
df_cleaned.to_csv('lf_clean_data_dpd.csv', index=False, encoding='utf-8')
```
