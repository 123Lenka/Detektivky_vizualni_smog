Vybrat sloupce do df
Zahodit otevirací dobu

```python
df = df[['name', 'city', 'zipcode', 'address', 'geolat', 'geolng']]

df.columns
df.head(2)
```

Ukázka a odstranění "( TEXT)" na konci řádků ve sloupci address a name

```python
adresy_se_zavorkou = df[df['address'].str.contains(r'\(.*\)$', na=False)]

adresy_se_zavorkou
```

df.loc[pro všechny řádky, ve sloupci address]

```python
df.loc[:, 'address'] = df['address'].str.replace(r'\s*\(.*\)$', '', regex=True)
adresy_se_zavorkou = df[df['address'].str.contains(r'\(.*\)$', na=False)]

df.columns
adresy_se_zavorkou
```

```python
df.loc[:, 'name'] = df['name'].str.replace(r'\s*\(.*\)$', '', regex=True)
name_se_zavorkou = df[df['name'].str.contains(r'\(.*\)$', na=False)]

name_se_zavorkou
```

Ukázka a oddělení č.p. z pole address

```python
adresy_s_cislem = df[df['address'].str.contains(r'\s\d[\dA-Za-z/,\- ]*(?:/[IVXLCDM]+)?$', na=False)]

df.columns
adresy_s_cislem
```

```python
adresy_bez_cisla = df[~df['address'].str.contains(r'\s\d[\dA-Za-z/,\- ]*(?:/[IVXLCDM]+)?$', na=False)]

df.columns
adresy_bez_cisla
```

extrahovat č.p. z address a poté vymazat z address

```python
df.loc[:,'house_no'] = df['address'].str.extract(r'(\d[\dA-Za-z/,\- ]*(/[IVXLCDM]+)?)$')[0]

df.head(2)
```

```python
df.loc[:,'address'] = df['address'].str.replace(r'\s\d[\dA-Za-z/,\- ]*(/[IVXLCDM]+)?$', '', regex = True)

df.head(2)
```

POMOCNÉ CSV
