```python
import pandas as pd

# Načtení dvou dim_box CSV souborů
path_1 = "C:/Users/Sabina/Sabi dokumenty/01 Czechitas/01 PYTHON/Python/Detektivky_vizualni_smog/sabi/data/dim_box_1.csv"
path_2 = "C:/Users/Sabina/Sabi dokumenty/01 Czechitas/01 PYTHON/Python/Detektivky_vizualni_smog/sabi/data/dim_box2.csv"

box1 = pd.read_csv(path_1, sep=";", encoding="utf-8", quotechar='"')
box2 = pd.read_csv(path_2, sep=";", encoding="utf-8", quotechar='"')

# # Sjednocení všech forem prázdných hodnot na pd.NA
# for d in [box1, box2]:
#     d.replace(["", "nan", "None"], pd.NA, inplace=True)

# Sjednocení všech forem prázdných hodnot na pd.NA
for d in [box1, box2]:
    d.replace({"": pd.NA, "nan": pd.NA, "None": pd.NA}, inplace=True)

# Sloučení obou datových rámců
df = pd.concat([box1, box2], ignore_index=True)

# Přegenerování unikátního ID
df['id'] = range(1, len(df) + 1)

# Oprava psc (odstranění .0)
df['psc'] = df['psc'].apply(lambda x: str(x).split('.')[0] if pd.notna(x) else pd.NA)

if "mesto.1" in df.columns:
    df = df.drop(columns=["mesto.1"])
    
# Oprava textových polí – nahrazení všech forem prázdnoty za ""
text_columns = ['cislo', 'psc', 'kod_obec', 'ulice', 'mesto']
for col in text_columns:
    df[col] = df[col].replace(["nan", "None", pd.NA], "").fillna("")
    df[col] = df[col].astype(str)

# # Převod typů
# df['cislo'] = df['cislo'].astype(str)
# df['psc'] = df['psc'].astype(str)
# df['kod_obec'] = df['kod_obec'].astype(str)
# df['ulice'] = df['ulice'].astype(str)
# df['mesto'] = df['mesto'].astype(str)
# df['lat'] = df['lat'].astype(float)
# df['long'] = df['long'].astype(float)

# # Převod pd.NA na prázdný řetězec a vše na string
# df = df.fillna("").astype(str)


# Export sjednoceného datasetu
df.to_csv(
    "C:/Users/Sabina/Sabi dokumenty/01 Czechitas/01 PYTHON/Python/Detektivky_vizualni_smog/sabi/data/dim_box.csv",
    sep=";",
    index=False,
    quoting=1,
    quotechar='"',
    encoding="utf-8"
)

print("Sloučený dataset s novými ID úspěšně uložen.")
```
