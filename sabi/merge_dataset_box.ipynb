```python
import pandas as pd

# Načti data
df_alza = pd.read_excel("Kopie souboru alzaboxes_enriched_ruiany.xlsx")
df_zasilkovna = pd.read_excel("Kopie souboru zasilkovna_enriched.xlsx")
df_gls = pd.read_excel("ss_clean_GLS_DATA.xlsx")

# Přidej označení zdroje
df_alza['source'] = 'alzabox'
df_zasilkovna['source'] = 'zasilkovna'
df_gls['source'] = 'gls'

# Sloučíme pod sebe
df_merged = pd.concat([df_alza, df_zasilkovna, df_gls], ignore_index=True)

# Vytvoříme nový unifikovaný ID sloupec (od 1)
df_merged['id'] = range(1, len(df_merged) + 1)

# Přesuneme 'id' na začátek (volitelné)
cols = ['id'] + [col for col in df_merged.columns if col != 'id']
df_merged = df_merged[cols]

# Ukázka výsledku
df_merged
```

```python
df = pd.DataFrame(df_merged)

df.head(2)
```

```python
df.columns
```

```python
df = df.drop(columns=['ID ', 'source'])
```

```python
df = df.rename(columns={
    'id': 'ID',
    'city_code.1': 'city'
  })
```

```python
df['zip_code'] = df['zip_code'].astype(str).str.replace(r'\.0$', '', regex=True)
```

```python
df['house_NO'] = df['house_NO'].astype(str).str.replace(r'\.0$', '', regex=True)
```

```python
df.to_csv('gls_alza_zasilkovna_dataset_clean.csv', index=False, encoding='utf-8')
```
